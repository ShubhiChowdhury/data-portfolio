```sql
CREATE OR REPLACE TABLE silver_layoffs AS
SELECT
  company,
  location,
  industry,
  country,
  stage,
  CAST(total_laid_off AS INT)           AS total_laid_off,
  CAST(percentage_laid_off AS DOUBLE)   AS pct_laid_off,
  CAST(funds_raised AS DOUBLE) AS funds_raised_millions,
  TO_DATE(date, 'yyyy-MM-dd')           AS layoff_date,
  YEAR(TO_DATE(date, 'yyyy-MM-dd'))     AS layoff_year,
  MONTH(TO_DATE(date, 'yyyy-MM-dd'))    AS layoff_month
FROM layoffs
WHERE total_laid_off IS NOT NULL;
```

```sql
SELECT * FROM silver_layoffs LIMIT 10;
```

```sql
CREATE OR REPLACE TABLE gold_layoffs_by_industry AS
SELECT industry, SUM(total_laid_off) AS total_laid_off
FROM silver_layoffs
WHERE industry IS NOT NULL
GROUP BY industry
ORDER BY total_laid_off DESC;
```

```sql
CREATE OR REPLACE TABLE gold_layoffs_by_country AS
SELECT country, SUM(total_laid_off) AS total_laid_off
FROM silver_layoffs
WHERE country IS NOT NULL
GROUP BY country
ORDER BY total_laid_off DESC;
```

```sql
CREATE OR REPLACE TABLE gold_layoffs_by_month AS
SELECT layoff_year, layoff_month, SUM(total_laid_off) AS total_laid_off
FROM silver_layoffs
GROUP BY layoff_year, layoff_month
ORDER BY layoff_year, layoff_month;
```

```sql
CREATE OR REPLACE TABLE gold_top_companies AS
SELECT company, SUM(total_laid_off) AS total_laid_off
FROM silver_layoffs
GROUP BY company
ORDER BY total_laid_off DESC
LIMIT 10;
```

```sql
CREATE OR REPLACE TABLE gold_layoffs_by_stage AS
SELECT stage, SUM(total_laid_off) AS total_laid_off
FROM silver_layoffs
WHERE stage IS NOT NULL AND stage != 'Unknown'
GROUP BY stage
ORDER BY total_laid_off DESC;
```

```sql
SELECT * FROM gold_layoffs_by_industry;
```

```sql
SELECT * FROM gold_layoffs_by_country;
```

```sql
SELECT CONCAT(layoff_year, '-', LPAD(layoff_month, 2, '0')) AS month_year,
       total_laid_off
FROM gold_layoffs_by_month
ORDER BY month_year;
```
